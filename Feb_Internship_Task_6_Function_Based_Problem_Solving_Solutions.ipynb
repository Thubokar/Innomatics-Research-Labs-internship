## Problem 1: Smart Parking Lot Management System
Design a function that processes vehicle IN/OUT logs, tracks parked vehicles, peak usage, and alerts on capacity.

```python
# Problem 1: Smart Parking Lot Management System

def smart_parking_manager(capacity, logs):
    """
    capacity: maximum parking capacity (int)
    logs: list of 'IN' / 'OUT' strings.
    Returns a dictionary with: current_parked, peak_usage, status.
    """
    current_parked = 0
    peak_usage = 0

    for entry in logs:
        if entry == 'IN':
            current_parked += 1
        elif entry == 'OUT' and current_parked > 0:
            current_parked -= 1

        # Check peak usage after current_parked has been updated for the entry
        if current_parked > peak_usage:
            peak_usage = current_parked

    # Determine parking status
    if current_parked > capacity:
        status = 'Over Capacity'
    elif current_parked == capacity:
        status = 'Full'
    else:
        status = 'Available'

    print(f'Currently Parked Vehicles: {current_parked}')
    print(f'Peak Usage: {peak_usage}')
    print(f'Parking Status: {status}')
    return {
        'current_parked': current_parked,
        'peak_usage': peak_usage,
        'status': status
    }

# Sample run using given sample input
smart_parking_manager(50, ['IN', 'IN', 'IN', 'OUT', 'IN', 'IN', 'OUT'])
```

## Problem 2: Online Food Delivery Time Estimator
Estimate delivery time based on distance, traffic level, and weather conditions.

```python
# Problem 2: Online Food Delivery Time Estimator

def estimate_delivery_time(distance_km, traffic_level, weather):
    """
    Base time assumed: 5 minutes fixed + 5 minutes per km.
    Traffic and weather add extra delay.
    traffic_level: 'Low', 'Medium', 'High'
    weather: 'Clear', 'Rainy', 'Stormy' (you can extend)
    """
    # Base time calculation
    base_time = 5 + distance_km * 5

    # Traffic delay
    traffic_level = traffic_level.lower()
    if traffic_level == 'low':
        traffic_delay = 0
    elif traffic_level == 'medium':
        traffic_delay = 5
    else:  # high
        traffic_delay = 10

    # Weather delay
    weather = weather.lower()
    if weather == 'clear':
        weather_delay = 0
    elif weather == 'rainy':
        weather_delay = 10
    else:  # stormy or severe
        weather_delay = 20

    eta = base_time + traffic_delay + weather_delay
    print(f'Estimated Delivery Time: {int(eta)} minutes')
    return eta

# Sample run to match example (approximate tuning)
estimate_delivery_time(8, 'High', 'Rainy')
```

## Problem 3: Movie Theatre Seat Occupancy Analyzer
Analyze booking data, calculate occupancy, and classify show status.

```python
# Problem 3: Movie Theatre Seat Occupancy Analyzer

def theatre_occupancy_analyzer(total_seats, booked_seats_list):
    """
    booked_seats_list can be a list of 1s representing booked seats.
    Calculates occupancy percentage and show status.
    Status rules (example):
    - 100% -> Housefull
    - >= 75% -> Almost Full
    - else -> Available
    """
    booked_count = 0
    for b in booked_seats_list:
        if b == 1:
            booked_count += 1

    occupancy_percent = (booked_count / total_seats) * 100 if total_seats > 0 else 0

    if occupancy_percent == 100:
        status = 'Housefull'
    elif occupancy_percent >= 75:
        status = 'Almost Full'
    else:
        status = 'Available'

    print(f'Occupancy: {int(occupancy_percent)}%')
    print(f'Show Status: {status}')
    return occupancy_percent, status

# Sample input: 150 booked out of 200 = 75%
booked_list = [1] * 150
theatre_occupancy_analyzer(200, booked_list)
```

## Problem 4: Cloud Server Load Classification System
Classify server status based on average CPU load readings.

```python
# Problem 4: Cloud Server Load Classification System

def classify_server_load(cpu_readings):
    """
    Accepts a list of CPU percentages, computes average, and
    classifies into Normal / Warning / Critical.
    """
    if not cpu_readings:
        avg = 0
    else:
        total = 0
        for val in cpu_readings:
            total += val
        avg = total / len(cpu_readings)

    if avg < 50:
        status = 'Normal'
    elif avg <= 80:
        status = 'Warning'
    else:
        status = 'Critical'

    print(f'Average CPU Load: {int(avg)}%')
    print(f'Server Status: {status}')
    return avg, status

# Sample run
classify_server_load([45, 60, 70, 85, 90])
```

## Problem 5: Smart Classroom Resource Usage Monitor
Track usage of resources and flag overused ones (e.g., over 8 hours).

```python
# Problem 5: Smart Classroom Resource Usage Monitor

def resource_usage_monitor(usage_dict, threshold=8):
    """
    usage_dict: dictionary of resource -> hours used.
    threshold: hours beyond which resource is considered overused.
    Prints overused resources and energy alert status.
    """
    overused = []
    for resource, hours in usage_dict.items():
        if hours > threshold:
            overused.append(resource)

    if overused:
        print('Overused Resources:', ', '.join(overused))
        print('Energy Alert: Yes')
    else:
        print('Overused Resources: None')
        print('Energy Alert: No')
    return overused

# Sample run
usage = {
    'Projector': 6,
    'AC': 9,
    'Lights': 4
}
resource_usage_monitor(usage)
```

## Problem 6: Online Event Registration Capacity Controller
Manage event registrations, prevent overbooking, and calculate waitlisted users.

```python
# Problem 6: Online Event Registration Capacity Controller

def event_registration_controller(capacity, registrations):
    """
    capacity: max allowed registrations.
    registrations: total attempted registrations (int).
    Prints confirmed, waitlisted, and status, returns them as dict.
    """
    if registrations <= capacity:
        confirmed = registrations
        waitlisted = 0
        status = 'Open'
    else:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = 'Closed'

    print(f'Confirmed Registrations: {confirmed}')
    print(f'Waitlisted Users: {waitlisted}')
    print(f'Registration Status: {status}')
    return {
        'confirmed': confirmed,
        'waitlisted': waitlisted,
        'status': status
    }

# Sample run
event_registration_controller(100, 105)
```

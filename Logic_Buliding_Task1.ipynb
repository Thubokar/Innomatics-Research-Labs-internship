1. User Login Check
Problem Statement:
Given a username and password, check whether login is successful.

username = "admin"
password = "1234"


```python
# Predefined correct credentials
USERNAME = "admin"
PASSWORD = "1234"

def check_login(input_username, input_password):
    """
    Check whether the given username and password
    match the stored credentials.
    Returns True if they match, otherwise False.
    """
    return input_username == USERNAME and input_password == PASSWORD

# Take username and password input from the user
user_username = input("Enter username: ")
user_password = input("Enter password: ")

# Check credentials using the function
if check_login(user_username, user_password):
    print("Login Successful")
else:
    print("Invalid Credentials")
```

 2. Pass / Fail Analyzer
Problem Statement:

Analyze student results.
marks = [45, 78, 90, 33, 60]


```python
# List of student marks
marks = [45, 78, 90, 33, 60]

# Initialize counters
pass_count = 0
fail_count = 0

# Go through each mark and update counters
for mark in marks:
    # A student passes if marks are greater than or equal to 50
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Print the final result clearly
print("Total students:", len(marks))
print("Passed students:", pass_count)
print("Failed students:", fail_count)
```

3.Simple Data Cleaner
Problem Statement:
Clean and standardize user names.

names = [" Alice ", "bob", " CHARLIE "]


```python
# Original list of names with inconsistent formatting
names = [" Alice ", "bob", " CHARLIE "]

# Clean and standardize names:
# 1. Remove extra spaces using strip()
# 2. Convert to lowercase using lower()
cleaned_names = []

for name in names:
    cleaned_name = name.strip().lower()   # remove spaces and convert to lowercase
    cleaned_names.append(cleaned_name)    # add cleaned name to new list

# Print the cleaned list
print("Cleaned names:", cleaned_names)
```

4. Message Length Analyzer
Problem Statement:
Analyze message sizes.

messages = ["Hi", "Welcome to the platform", "OK"]


```python
# List of messages
messages = ["Hi", "Welcome to the platform", "OK"]

# Analyze each message
for msg in messages:
    length = len(msg)  # Get length of the message
    print(f"Message: '{msg}' | Length: {length}", end="")

    # Flag messages longer than 10 characters
    if length > 10:
        print(" -> Long message")
    else:
        print(" -> Short message")
```

5. Error Message Detector
Problem Statement:
Detect error messages from system logs.

logs = ["INFO", "ERROR", "WARNING", "ERROR"]


```python
# System logs
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Initialize error counter
error_count = 0

# Check each log entry
for log in logs:
    if log == "ERROR":
        error_count += 1  # Increment counter for each ERROR found

# Print the total error count
print("Total ERROR messages:", error_count)
```

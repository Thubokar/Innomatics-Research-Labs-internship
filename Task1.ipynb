```python
# Predefined correct credentials
USERNAME = "admin"
PASSWORD = "1234"

def check_login(input_username, input_password):
    """
    Check whether the given username and password
    match the stored credentials.
    Returns True if they match, otherwise False.
    """
    return input_username == USERNAME and input_password == PASSWORD

# Take username and password input from the user
user_username = input("Enter username: ")
user_password = input("Enter password: ")

# Check credentials using the function
if check_login(user_username, user_password):
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# List of student marks
marks = [45, 78, 90, 33, 60]

# Initialize counters
pass_count = 0
fail_count = 0

# Go through each mark and update counters
for mark in marks:
    # A student passes if marks are greater than or equal to 50
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Print the final result clearly
print("Total students:", len(marks))
print("Passed students:", pass_count)
print("Failed students:", fail_count)
```

## Problem 1: Unique Words in a Sentence

**Problem Statement:** Given a sentence, identify all unique words and display:
1. The count of unique words
2. The unique words themselves


```python
# Problem 1: Unique Words in a Sentence

# Input sentence
sentence = "python is easy and python is powerful"

# Split the sentence into individual words
words = sentence.split()

# Convert the list of words into a set to remove duplicates
unique_words = set(words)

# Display results
print(f"Unique words count: {len(unique_words)}")
print(f"Unique words: {unique_words}")
```

## Problem 2: Highest Salary from Employee Data

**Problem Statement:** Store employee names and their salaries, and determine which employee has the highest salary.



```python
# Problem 2: Highest Salary from Employee Data

# Employee dictionary with names as keys and salaries as values
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find the employee with the highest salary using max() function
max_employee = max(employees, key=employees.get)
max_salary = employees[max_employee]

# Display result
print(f"Highest Salary: {max_employee} - {max_salary}")
```

## Problem 3: Find Maximum and Minimum Values

**Problem Statement:** Find the maximum and minimum values from a list of numbers.



```python
# Problem 3: Find Maximum and Minimum Values

# List of numbers
numbers = [45, 22, 89, 10, 66]

# Initialize both maximum and minimum with the first element
maximum = numbers[0]
minimum = numbers[0]

# Loop through each element and compare
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

# Display results
print(f"List: {numbers}")
print(f"Max: {maximum}")
print(f"Min: {minimum}")
```

## Problem 4: Count Products Above a Price Threshold

**Problem Statement:** Count how many products have a price greater than 1000.



```python
# Problem 4: Count Products Above a Price Threshold

# List of product prices
prices = [450, 1200, 899, 1500, 300]

# Initialize counter
count = 0

# Loop through prices and compare with threshold (1000)
for price in prices:
    if price > 1000:
        count += 1

# Display result
print(f"Products above 1000: {count}")
```

## Problem 5: Calculate Attendance Percentage

**Problem Statement:** Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).



```python
# Problem 5: Calculate Attendance Percentage

# Attendance list with Present (P) and Absent (A)
attendance = ["P", "P", "A", "P", "P"]

# Count the number of present days
present_count = attendance.count("P")

# Calculate total days
total_days = len(attendance)

# Calculate attendance percentage
attendance_percentage = (present_count / total_days) * 100

# Display result
print(f"Attendance Percentage: {attendance_percentage}")
```

## Problem 7: Count Character Frequency

**Problem Statement:** Count how many times each character appears in a given string.



```python
# Problem 7: Count Character Frequency

# Input string
text = "python"

# Initialize empty dictionary to store character frequency
char_frequency = {}

# Loop through each character and count occurrences
for char in text:
    if char in char_frequency:
        char_frequency[char] += 1  # Increment count if character exists
    else:
        char_frequency[char] = 1   # Initialize count if character is new

# Display result
print(char_frequency)
```

## Problem 9: Check if a Key Exists in a Dictionary

**Problem Statement:** Check whether a specific key exists in a dictionary.



```python
# Problem 9: Check if a Key Exists in a Dictionary

# Employee dictionary
employees = {
    "John": "Manager",
    "Sarah": "Developer",
    "Mike": "Analyst"
}

# Key to check
key_to_check = "John"

# Check if key exists using 'in' keyword
if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

## Problem 10: Calculate Average Marks

**Problem Statement:** Calculate the average marks of a student from a list of marks.



```python
# Problem 10: Calculate Average Marks

# List of marks
marks = [85, 90, 60, 55]

# Calculate sum of all marks
total_marks = sum(marks)

# Calculate average by dividing sum by total number of marks
average_marks = total_marks / len(marks)

# Display result
print(f"Average Marks: {average_marks}")
```

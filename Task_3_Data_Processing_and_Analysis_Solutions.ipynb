## Problem 1: Employee Performance Bonus Eligibility

Given a dictionary of employee performance scores, identify the highest score, handle ties, and display all employees eligible for the top performance bonus.

```python
# Problem 1: Employee Performance Bonus Eligibility

# Dictionary of employees and their performance scores
employees = {
    'Ravi': 92,
    'Anita': 88,
    'Kiran': 92,
    'Suresh': 85
}

# Step 1: Find the highest performance score
highest_score = max(employees.values())

# Step 2: Collect all employees who have this highest score (handle ties)
top_performers = [name for name, score in employees.items() if score == highest_score]

# Step 3: Display result in required format
names_str = ', '.join(top_performers)
print(f'Top Performers Eligible for Bonus: {names_str} (Score: {highest_score})')
```

## Problem 2: Search Query Keyword Analysis

Convert the search query to lowercase, ignore punctuation, count frequency of each keyword, and display only keywords searched more than once.

```python
# Problem 2: Search Query Keyword Analysis

import string

# Input search query
query = 'Buy mobile phone buy phone online'

# Step 1: Convert to lowercase
query = query.lower()

# Step 2: Remove common punctuation characters
translator = str.maketrans('', '', string.punctuation)
clean_query = query.translate(translator)

# Step 3: Split into individual keywords
words = clean_query.split()

# Step 4: Count frequency of each keyword using a dictionary
freq = {}
for word in words:
    if word in freq:
        freq[word] += 1
    else:
        freq[word] = 1

# Step 5: Filter and display keywords that appear more than once
repeated_keywords = {word: count for word, count in freq.items() if count > 1}
print(repeated_keywords)
```

## Problem 3: Sensor Data Validation

Identify even sensor readings, treat them as valid, and store them as (hour_index, reading_value) pairs. Ignore odd readings.

```python
# Problem 3: Sensor Data Validation

# List of hourly sensor readings
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Step 1: Initialize empty list to store valid (hour, value) pairs
valid_readings = []

# Step 2: Iterate through readings with their index (hour)
for hour, value in enumerate(sensor_readings):
    if value % 2 == 0:  # even number = valid reading
        valid_readings.append((hour, value))

# Step 3: Display valid sensor readings
print('Valid Sensor Readings (Hour, Value):')
print(valid_readings)
```

## Problem 4: Email Domain Usage Analysis

Count how many users belong to each email domain and calculate percentage usage of each domain.

```python
# Problem 4: Email Domain Usage Analysis

# List of employee email IDs
emails = [
    'ravi@gmail.com',
    'anita@yahoo.com',
    'kiran@gmail.com',
    'suresh@gmail.com',
    'meena@yahoo.com'
]

# Step 1: Extract domains and count occurrences
domain_counts = {}
for email in emails:
    domain = email.split('@')[1]
    if domain in domain_counts:
        domain_counts[domain] += 1
    else:
        domain_counts[domain] = 1

# Step 2: Calculate total number of emails
total_emails = len(emails)

# Step 3: Calculate and print percentage usage for each domain
for domain, count in domain_counts.items():
    percentage = (count / total_emails) * 100
    print(f'{domain}: {int(percentage)}%')
```

## Problem 5: Sales Spike Detection

Calculate the average daily sales and detect days where sales are more than 30% above average. Display the day number and sale value.

```python
# Problem 5: Sales Spike Detection

# List of daily sales values
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculate average sales
average_sales = sum(sales) / len(sales)

# Step 2: Calculate threshold for spike (30% above average)
threshold = average_sales * 1.3

# Step 3: Detect days with sales above threshold
for day_index, sale in enumerate(sales, start=1):  # days are 1-based
    if sale > threshold:
        print(f'Day {day_index}: {sale}')
```

## Problem 6: Duplicate User ID Detection

Identify duplicate user IDs and display how many times each duplicate appears.

```python
# Problem 6: Duplicate User ID Detection

# List of user IDs
user_ids = ['user1', 'user2', 'user1', 'user3', 'user1', 'user3']

# Step 1: Count occurrences of each user ID
id_counts = {}
for uid in user_ids:
    if uid in id_counts:
        id_counts[uid] += 1
    else:
        id_counts[uid] = 1

# Step 2: Filter only duplicates (count > 1) and display
for uid, count in id_counts.items():
    if count > 1:
        print(f'{uid} → {count} times')
```

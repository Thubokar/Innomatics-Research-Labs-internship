## Problem 1: Mobile Recharge Validation System
Create a function to validate a mobile recharge request with retry using a while loop.

```python
# Problem 1: Mobile Recharge Validation System

VALID_PLANS = [199, 299, 399, 599]

def validate_recharge(amount):
    """Return True if amount is valid, else False.
    Conditions:
    - amount must be at least 50
    - if using a plan, it should be one of VALID_PLANS
    """
    if amount < 50:
        return False
    if amount in VALID_PLANS or amount >= 50:
        return True
    return False

# Retry loop for user input (interactive usage)


while True:
     try:
         amt = int(input('Enter recharge amount (₹): '))
     except ValueError:
         print('Invalid input. Please enter a numeric amount.')
         continue

     if validate_recharge(amt):
         print(f'Recharge successful for ₹{amt}.')
         break
     else:
         print('Invalid recharge amount. Try again (min ₹50, or valid plans 199/299/399/599).')
```

## Problem 2: Inventory Reorder Alert System
Given a dictionary of products and stock quantities, print alerts using a function and a for loop.

```python
# Problem 2: Inventory Reorder Alert System

def check_inventory(stock_dict):
    """Print Reorder Alert if stock < 15, else Stock OK for each product."""
    for product, qty in stock_dict.items():
        if qty < 15:
            print(f'{product}: Reorder Alert (Stock = {qty})')
        else:
            print(f'{product}: Stock OK (Stock = {qty})')

# Example usage
inventory = {
    'Apples': 10,
    'Bananas': 25,
    'Oranges': 14,
    'Grapes': 30
}

check_inventory(inventory)
```

## Problem 3: Student Result Processing System
Create a function that accepts a list of marks, calculates average using a loop, and prints Pass/Fail.

```python
# Problem 3: Student Result Processing System

def process_result(marks):
    """Accepts list of marks, calculates average, and returns 'Pass' or 'Fail'."""
    total = 0
    for m in marks:
        total += m
    avg = total / len(marks) if marks else 0

    print(f'Average Marks: {avg}')
    if avg >= 50:
        print('Pass')
        return 'Pass'
    else:
        print('Fail')
        return 'Fail'

# Example usage
student_marks = [60, 55, 70, 45]
process_result(student_marks)
```

## Problem 4: Cab Fare Estimator with Retry Option
Create a cab fare calculator function and allow retry using a while loop.

```python
# Problem 4: Cab Fare Estimator with Retry Option

BASE_FARE = 50
PER_KM = 12
PEAK_EXTRA = 0.25  # 25% extra

def calculate_fare(distance_km, is_peak_hour):
    """Calculate cab fare based on distance and peak hour flag.
    Base fare = 50, ₹12 per km, 25% extra during peak hours."""
    fare = BASE_FARE + (PER_KM * distance_km)
    if is_peak_hour:
        fare = fare * (1 + PEAK_EXTRA)
    return fare

# Retry loop for interactive usage


while True:
    try:
        dist = float(input('Enter distance in km: '))
    except ValueError:
        print('Invalid distance. Please enter a number.')
        continue

    peak_input = input('Is it peak hour? (yes/no): ').strip().lower()
    is_peak = peak_input == 'yes'

    fare = calculate_fare(dist, is_peak)
    print(f'Total Fare: ₹{fare:.2f}')

    retry = input('Do you want to calculate again? (yes/no): ').strip().lower()
    if retry != 'yes':
        break
```

## Problem 5: Employee Attendance Eligibility Checker
Create a function that calculates attendance percentage and returns eligibility.

```python
# Problem 5: Employee Attendance Eligibility Checker

def check_attendance(attendance_list):
    """
    Accepts a list with 'P' (Present) and 'A' (Absent).
    Calculates attendance percentage and
    returns 'Eligible' if ≥ 75%, else 'Not Eligible'.
    """
    total_days = len(attendance_list)
    present_days = 0
    for status in attendance_list:
        if status == 'P':
            present_days += 1
    percentage = (present_days / total_days) * 100 if total_days > 0 else 0
    print(f'Attendance Percentage: {percentage:.2f}%')
    if percentage >= 75:
        return 'Eligible'
    else:
        return 'Not Eligible'

# Example usage
attendance = ['P', 'P', 'A', 'P', 'P', 'A', 'P', 'P']
result = check_attendance(attendance)
print(result)
```

## Problem 6: Password Strength Checker
Create a function that checks password length, presence of at least one digit, and one special character (@#$).

```python
# Problem 6: Password Strength Checker

def is_strong_password(password):
    """
    Check if password is strong based on rules:
    - Minimum length 8
    - At least one digit
    - At least one special character from @#$
    Returns True if strong, else False.
    """
    if len(password) < 8:
        return False

    has_digit = False
    has_special = False
    specials = '@#$'

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in specials:
            has_special = True

    return has_digit and has_special

# Example usage
test_passwords = ['abc123', 'password1', 'Pass@123', 'weakpass', 'Strong#9']
for pwd in test_passwords:
    print(pwd, '->', 'Strong' if is_strong_password(pwd) else 'Weak')
```

## Problem 1: Social Media – Post Engagement Analyzer
Create a function to calculate total likes and classify the post as Viral or Normal Engagement.

```python
# Problem 1: Social Media – Post Engagement Analyzer

def analyze_post_engagement(likes_list):
    """
    Accepts a list of like counts (e.g., from different time intervals),
    calculates total likes using a loop, and returns
    (total_likes, status) where status is 'Viral Post' or 'Normal Engagement'.
    """
    total_likes = 0
    for likes in likes_list:
        total_likes += likes

    if total_likes >= 1000:
        status = 'Viral Post'
    else:
        status = 'Normal Engagement'
    return total_likes, status

# Example usage
likes_data = [200, 350, 500]  # total = 1050
total, state = analyze_post_engagement(likes_data)
print(f'Total Likes: {total}')
print(f'Post Status: {state}')
```

## Problem 3: Agriculture – Rainfall Adequacy Checker
Create a function to compute average rainfall and compare with required level.

```python
# Problem 3: Agriculture – Rainfall Adequacy Checker

def check_rainfall_adequacy(rainfall_list, required_level):
    """
    Accepts list of rainfall values and required_level.
    Uses a loop to calculate average and returns (average, status).
    """
    if not rainfall_list:
        return 0, 'Inadequate Rainfall'

    total = 0
    for r in rainfall_list:
        total += r
    avg = total / len(rainfall_list)

    if avg >= required_level:
        status = 'Adequate Rainfall'
    else:
        status = 'Inadequate Rainfall'

    print(f'Average Rainfall: {avg}')
    print(f'Rainfall Status: {status}')
    return avg, status

# Example usage
rainfall_data = [60, 80, 76]  # avg = 72
check_rainfall_adequacy(rainfall_data, required_level=70)
```

## Problem 4: Social Media – Duplicate Account Detection
Create a function to detect duplicate usernames using a set.

```python
# Problem 4: Social Media – Duplicate Account Detection

def has_duplicate_accounts(usernames):
    """
    Accepts a list of usernames.
    Returns True if duplicates exist, else False.
    Also prints 'Duplicate Accounts Found: Yes/No'.
    """
    unique_users = set(usernames)
    duplicates_exist = len(unique_users) != len(usernames)
    print('Duplicate Accounts Found:', 'Yes' if duplicates_exist else 'No')
    return duplicates_exist

# Example usage
user_list = ['userA', 'userB', 'userC', 'userA']
has_duplicate_accounts(user_list)
```

## Problem 6: Agriculture – Premium Crop Price Filter
Create a function that filters prices above the premium threshold (₹2000).

```python
# Problem 6: Agriculture – Premium Crop Price Filter

def filter_premium_crops(prices, threshold=2000):
    """
    Accepts list of crop prices and threshold.
    Returns list of premium crop prices (> threshold).
    """
    premium = []
    for price in prices:
        if price > threshold:
            premium.append(price)
    print(f'Premium Crops: {premium}')
    return premium

# Example usage
crop_prices = [1500, 2500, 1800, 3200]
filter_premium_crops(crop_prices)
```

## Problem 7: System Monitoring – Application Health Checker
Create a function to classify system health based on error count.

```python
# Problem 7: System Monitoring – Application Health Checker

def application_health_status(error_count):
    """
    Classify system health based on error_count:
    0       -> Healthy
    1 to 5 -> Minor Issues
    >5     -> Critical Issues
    """
    print(f'Error Count: {error_count}')
    if error_count == 0:
        status = 'Healthy'
    elif error_count <= 5:
        status = 'Minor Issues'
    else:
        status = 'Critical Issues'
    print(f'System Status: {status}')
    return status

# Example usage
application_health_status(7)
```

## Problem 9: E-Learning – Student Attendance Eligibility System
Create a function to calculate attendance percentage from a list and decide exam eligibility.

```python
# Problem 9: E-Learning – Student Attendance Eligibility System

def check_exam_eligibility(attendance_list):
    """
    Accepts list with 'P' and 'A', calculates percentage using a loop,
    prints percentage and eligibility, and returns (percentage, status).
    """
    total_days = len(attendance_list)
    present_days = 0
    for status in attendance_list:
        if status == 'P':
            present_days += 1
    percentage = (present_days / total_days) * 100 if total_days > 0 else 0
    if percentage >= 75:
        eligibility = 'Eligible'
    else:
        eligibility = 'Not Eligible'
    print(f'Attendance Percentage: {percentage}')
    print(f'Exam Eligibility: {eligibility}')
    return percentage, eligibility

# Example usage
attendance_data = ['P', 'P', 'A', 'P', 'P']  # 4/5 = 80%
check_exam_eligibility(attendance_data)
```

## Problem 10: Smart Electricity Bill Analyzer
Create a function to calculate the bill using slabs and classify usage level.

```python
# Problem 10: Smart Electricity Bill Analyzer

def calculate_electricity_bill(units):
    """
    Calculate electricity bill based on units consumed using slabs:
    - First 100 units  -> ₹3 per unit
    - Next 100 units   -> ₹5 per unit (101–200)
    - Above 200 units  -> ₹7 per unit

    Also classify usage level based on total bill:
    - bill < 500         -> 'Low Usage'
    - 500 to 1500       -> 'Moderate Usage'
    - > 1500            -> 'High Usage'

    Returns (bill, usage_status).
    """
    bill = 0
    remaining_units = units

    # First 100 units
    first_slab = min(remaining_units, 100)
    bill += first_slab * 3
    remaining_units -= first_slab

    # Next 100 units (101–200)
    if remaining_units > 0:
        second_slab = min(remaining_units, 100)
        bill += second_slab * 5
        remaining_units -= second_slab

    # Above 200 units
    if remaining_units > 0:
        bill += remaining_units * 7

    # Classify usage based on bill amount
    if bill < 500:
        usage_status = 'Low Usage'
    elif bill <= 1500:
        usage_status = 'Moderate Usage'
    else:
        usage_status = 'High Usage'

    print(f'Total Units: {units}')
    print(f'Total Bill: ₹{bill}')
    print(f'Usage Status: {usage_status}')
    return bill, usage_status

# Example usage
calculate_electricity_bill(250)
```

## Problem 1: Smart Light Controller
Create a class to control a smart light that can be turned ON or OFF and display its status.

```python
# Problem 1: Smart Light Controller

class SmartLight:
    def __init__(self, name):
        self.name = name
        self.is_on = False

    def turn_on(self):
        self.is_on = True

    def turn_off(self):
        self.is_on = False

    def display_status(self):
        status = 'ON' if self.is_on else 'OFF'
        print(f'{self.name} is {status}')

# Sample usage
light = SmartLight('Bedroom Light')
light.turn_on()
light.display_status()
```

## Problem 2: Employee ID Card System
Create a class to store employee name, ID, and department and display ID card details.

```python
# Problem 2: Employee ID Card System

class EmployeeIDCard:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_card(self):
        print('Employee ID Card')
        print(f'Name: {self.name}')
        print(f'ID: {self.emp_id}')
        print(f'Department: {self.department}')

# Sample usage
emp = EmployeeIDCard('Rahul', 'EMP102', 'IT')
emp.display_card()
```

## Problem 3: Mobile Contact Record
Create a class to store contact name and phone number and display contact information.

```python
# Problem 3: Mobile Contact Record

class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print('Contact Saved')
        print(f'Name: {self.name}')
        print(f'Phone: {self.phone}')

# Sample usage
contact = Contact('Anita', 9876543210)
contact.display_contact()
```

## Problem 4: Product Price Tag Generator
Create a class that stores product name and price and prints a formatted price tag.

```python
# Problem 4: Product Price Tag Generator

class ProductPriceTag:
    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price = price

    def print_price_tag(self):
        print(f'Product: {self.product_name}')
        print(f'Price: ₹{self.price}')

# Sample usage
product = ProductPriceTag('Headphones', 2499)
product.print_price_tag()
```

## Problem 5: Movie Rating Display System
Create a class that stores movie name and rating and displays movie details.

```python
# Problem 5: Movie Rating Display System

class MovieRating:
    def __init__(self, movie_name, rating):
        self.movie_name = movie_name
        self.rating = rating

    def display_movie_details(self):
        print(f'Movie: {self.movie_name}')
        print(f'Rating: {self.rating} / 5')

# Sample usage
movie = MovieRating('Inception', 4.8)
movie.display_movie_details()
```

## Problem 6: Delivery Address Manager
Create a class that stores customer name and delivery address and prints delivery details.

```python
# Problem 6: Delivery Address Manager

class DeliveryAddressManager:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def print_delivery_details(self):
        print('Delivery Details')
        print(f'Customer: {self.customer_name}')
        print(f'Address: {self.address}')

# Sample usage
delivery = DeliveryAddressManager('Suman', 'Hyderabad')
delivery.print_delivery_details()
```
